## Reshaping, stacking, squeezing, unsqueezing

* reshaping- reshapes an inout tensor to a defined shape
* view = return a view of an input tensor of certain shape but keeps the same memeory
stacking - vstack, hstack
* squeeze - remoce all '1' dimesions from a tensor
* unsqueeze - add '1' dimoersion to atarget tensor
* permute - return a view of the input with dimensrions permutatted

```python
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
x_reshaped = x.reshape(9, 1)
x_reshaped, x_reshaped.shape
```

```python
# change the view
```

```python
z = x.view(1,9)
z, z.shape
```

```python
z[0][0] = 5
```

```python
z
```

```python
x_stacked = torch.stack([x, x, x, x], dim = 1)
x_stacked
```

```python
 x_reshaped.shape
```

```python
x_reshaped.unsqueeze(1), x_reshaped.unsqueeze(1).shape
```

```python
x_original = torch.rand(size = (225, 225, 3))
x_permute = x_original.permute(2, 0, 1)

print(f"previos shape: {x_original.shape}")
print(f"new shape: {x_permute.shape}")
```

```python
x_original[0, 0, 0] = 122.5
```

```python
x_permute[0,0,0]
```

## Reproducibilty

```python
import torch
random_tensor1 = torch.rand(3,4)
random_tensor2 = torch.rand(3,4)

print(random_tensor1)
print(random_tensor2)
print(random_tensor1 == random_tensor2)
```

```python
RANDOM_STATE = 42
torch.manual_seed(RANDOM_STATE)

random_tensor3 = torch.rand(3,4)

torch.manual_seed(RANDOM_STATE)
random_tensor4 = torch.rand(3,4)

print(random_tensor3)
print(random_tensor4)
print(random_tensor3 == random_tensor4)
```
